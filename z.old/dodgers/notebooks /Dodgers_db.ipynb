## Game Logs

```python
game_logs = pd.read_sql_query("SELECT * FROM game_logs", conn)

game_logs.columns
```

```python
game_logs.head(5)
```

```python
#Setting up the new dataframe
player_game_logs = pd.read_sql_query("SELECT players.name, game_logs.* FROM players JOIN game_logs On players.mlb_player_id = game_logs.mlb_player_id ", conn)

#rename the columns

player_game_logs.rename(columns={
    "mlb_player_id": "Player ID",
    "game_date": "Game Date",
    "team": "Team",
    "opponent": "Opponent",
    "ab": "At Bats",
    "r": "Runs",
    "h": "Hits",
    "tb": "Total Bases",
    "doubles": "Doubles",
    "triples": "Triples",
    "hr": "Home Runs",
    "rbi": "RBIs",
    "bb": "Walks",
    "ibb": "Intentional Walks",
    "so": "Strikeouts",
    "sb": "Stolen Bases",
    "cs": "Caught Stealing",
    "avg": "Batting Average",
    "obp": "On Base %",
    "slg": "Slugging %",
    "hbp": "Hit By Pitch",
    "sac": "Sac Bunts",
    "sf": "Sac Flies"
}, inplace=True)

print("Collumns renamed successfully!")
player_game_logs.columns
```
